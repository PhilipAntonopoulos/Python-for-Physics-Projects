When two particles collide they act as a _resonating system_ because of the probability distributions changing. When that happens the chances of a collision happening increases and then the two particles decay into more and then collide exiting the system, but leaving their decay products to be detected. The chance of such collisions to happen can be modeled with such a distribution.



* With a given mass of M = 2.5, plot the distributions with $\Gamma$ = 1,2,3

```python
import numpy as np
import matplotlib.pyplot as plt




def Breit_Wigner(E,M,G):
    prob = (M**3  * G)/(( (E**2) - (M**2) )**2 + ( (M**2) * (G**2)) )
    return prob


x = np.arange(2, 5, 0.01)

plt.plot(x,Breit_Wigner(x,2.5,1), color = 'blue', label = 'gamma = 1')
plt.plot(x,Breit_Wigner(x,2.5,2), color = "red", label = 'gamma = 2')
plt.plot(x,Breit_Wigner(x,2.5,3), color = 'orange', label = 'gamma = 3')

plt.xlabel('Energy')
plt.ylabel('P(E)')
plt.legend()
```
