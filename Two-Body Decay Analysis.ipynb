The task at hand is to find the mass of particle c. We are given the masses of the initial particle and particle b. We are also given experimental values of the momentum of the b particle. <br> 

* To solve this problem we are going to use the two elemental conservetion laws; the one of energy and the one of momentum. <br>
* Therefore $P_{before} = P_{After}$ (1) and $ E_{before} = E_{After} $ (2)
* From the special theory of relativity, we know that the the energy of a particle is dependent on its mass and momentum, as expressed in Einstein's formula $ E = \sqrt{m^2 + p^2} $ , working in units where c = 1.
* We also know that the momentum of a particle due to relativistic effects, undergoes a Loretz transform, therefore is $p = \gamma m v $ , where $ \gamma $ is equal to $ \frac{1}{\sqrt{1 - v^2}}$(for c = 1), rather than being equal to $ p = mv $.
* From (1) and (2) and the above formulas, it is trivial to find that $ m_c = \sqrt{m_A^2 - 2m_A\sqrt{m_B^2 + p_B^2} + m_B^2}$
    

### Proof

Using $P_{before} = P_{After} \iff p_C = -p_B$



Therefore:



So, the final formula used is $ m_c = \sqrt{5 - 4\sqrt{1 + p_B^2}}$

```python
# Importing the momenta for the b particle

import numpy as np
import matplotlib.pyplot as plt

filename = '2body.txt'
ps = np.loadtxt(filename)

#Plotting the distribution of the momenta

plt.hist(ps, bins = 50)
plt.xlabel("Momentum of mass b (eV / c)")
plt.title("Momentum Distribution of mass b")


```

```python
import numpy as np
import matplotlib.pyplot as plt

def missing_mass(Pb):
    m_c = np.sqrt(5 - 4* np.sqrt(1 + (Pb **2)))
    return m_c

m = missing_mass(ps)
std = np.std(m)
mn = np. mean(m)

plt.hist(m, bins = 50)

plt.axvline(mn, color = 'red')

plt.axvline(mn + std, linestyle = 'dashed', color = 'red')

plt.axvline(mn - std, linestyle = 'dashed', color = 'red')

print("The mean is: ", mn)
print("The standard deviation is: ", std)

plt.xlabel("Missing Mass of particle c(eV / c²)")
plt.title("Mass of Particle c Distribution")
```
