* Write code to find the derivative function of a function f at a non-zero x, distance a away from x.

```python


def derivative(f, x, a):
    numer = f(x+a) - f(x-a)
    denom = 2*a
    if denom != 0:
        deriv = numer / denom
        return deriv
    else:
        return 'null'
    
def f(x): return x **2

derivative(f, 3, 0.01)
    
```

* Varying a = $ [10^{-5}, 0.1) $ and fixing x = 1. Calculate the numerical derivative of the following functions: $e^x, log(x), sin(x), cos(x), \frac{x^3}{6} + \frac{x^4}{8}$ and plot it and the error as a function of a.



```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

fig , (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2)

def derivative(f, x, a):
    numer = f(x+a) - f(x-a)
    denom = 2*a
    if denom.all() != 0:
        deriv = numer / denom
        return deriv
    else:
        return 'null'
    
a = np.linspace(10** -5 , 0.1, 20)

def e(x):return np.sin(x)

ax1.plot(a,derivative(e, 0, a))

ax1.set_xlabel("a")
ax1.set_ylabel(" Numerical Derivative of $e^x$ ")

delta = 0.1

ax2.plot(a, (100/ delta) * (abs(derivative(e,0,a) - derivative(e,0,a+delta))))

ax2.set_xlabel("a")
ax2.set_ylabel(" $\sigma _ a$ ")

    
    
```

* With numerical derivative function at hand, plot the derivative of the aforementioned function as a function of x. Let a = $10^{-3}$ 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

fig , (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2)

def derivative(f, x, a):
    numer = f(x+a) - f(x-a)
    denom = 2*a
    deriv = numer / denom
    return deriv
    
a = 10 ** -3

def f(x):return ((x**3 / 6) + (x**4 / 8))

x = np.linspace(0.1,5, 60)

ax1.plot(x, f(x))

ax2.plot(x, derivative(f,x,a))
```
